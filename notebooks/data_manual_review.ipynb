```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../nifty100.db")

pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table'
""", conn)
```

5 RANDOM COMPANIES

```python
pd.read_sql("""
SELECT id
FROM companies
ORDER BY RANDOM()
LIMIT 5
""", conn)
```

YEAR COVERAGE

```python
pd.read_sql("""
SELECT
    company_id,
    COUNT(*) AS pl_records,
    MIN(year) AS first_year,
    MAX(year) AS last_year
FROM profitandloss
WHERE company_id IN (
    'HEROMOTOCO',
    'TATAMOTORS',
    'NHPC',
    'ATGL',
    'TRENT'
)
GROUP BY company_id
ORDER BY company_id
""", conn)
```

BALANCE SHEET COVERAGE

```python
pd.read_sql("""
SELECT
    company_id,
    COUNT(*) AS bs_records,
    MIN(year) AS first_year,
    MAX(year) AS last_year
FROM balancesheet
WHERE company_id IN (
    'HEROMOTOCO',
    'TATAMOTORS',
    'NHPC',
    'ATGL',
    'TRENT'
)
GROUP BY company_id
ORDER BY company_id
""", conn)
```

CASH FLOW COVERAGE

```python
pd.read_sql("""
SELECT
    company_id,
    COUNT(*) AS cf_records,
    MIN(year) AS first_year,
    MAX(year) AS last_year
FROM cashflow
WHERE company_id IN (
    'HEROMOTOCO',
    'TATAMOTORS',
    'NHPC',
    'ATGL',
    'TRENT'
)
GROUP BY company_id
ORDER BY company_id
""", conn)
```

```python
pd.read_sql("""
SELECT *
FROM cashflow
WHERE company_id = 'ATGL'
""", conn)
```

```python
pd.read_sql("""
SELECT
    company_id,
    COUNT(*) AS years_available
FROM profitandloss
GROUP BY company_id
HAVING COUNT(*) < 5
ORDER BY years_available
""", conn)
```

```python
pd.read_sql("""
SELECT id
FROM companies
WHERE id = 'ATGL'
""", conn)
```

```python
pd.read_sql("""
SELECT COUNT(*) AS cashflow_rows
FROM cashflow
""", conn)
```

```python
pd.read_sql("""
SELECT DISTINCT company_id
FROM cashflow
ORDER BY company_id
""", conn)
```
